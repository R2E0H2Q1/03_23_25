```python
import numpy as np
from IPython.display import display
import pandas as pd


print('Homework PD4')
print('Questions - Read the following csv file into a data frame:')

df = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/23.03.2025/refs/heads/main/mobile.csv")
display('Original Table', df)
print()
```

```python
#1. Create a first name column from the first word of the subscriber's name, use the apply command

def first_name(name):
    return str(name).split()[0]
    print(first_name())

df['First Name'] = df['Customer Name'].apply(first_name)
display('1.Adding new column: First Name', df)
```

```python
#2 Create a column for the cellular company by phone number prefix: 050 Pelephone, 051 Golan, 052 Cellcom, 053 Motorola
#(The company name will appear in English, for example Pelephone) Hint: replace

prefix = {
    '050': 'Pelephone',
    '051': 'Golan',
    '052': 'Cellcom',
    '053': 'Motorola',
    '054': 'Hotmobile'
}

df['Company'] = df['Phone Number'].str[:3].replace(prefix)

display('2. New Column-Company:', df)
```

```python
# 3 Use the column you created in the previous section and calculate:

# a. How many subscribers does each cell phone company have? Hint: counts value

display('3.a. Amount of Subscribers: ', df['Company'].value_counts())
print()

# b. What are the names of the different cell phone companies? Hint: unique

display('3.b. Company names: ', df['Company'].unique())
print()

# c. How many different cell phone companies are there? Hint: nunique

display('3.c. How many companies: ', df['Company'].nunique())
```

```python
# 4 **Authority: Create a column that will write the user type:
# If the talk time is greater than the surfing time, the value talker will appear, otherwise the value surfer will appear, solve in 2 ways:

# a. apply lambda row 

df['User Type'] = df.apply(lambda row: 'Talker' if row['Talked'] > row['Internet'] else 'Surfer', axis=1)
display('4.a. Using Lambda: ', df)


print()

# b. vectorize 

def user_type(talked, internet):
    if talked > internet:
        return 'Talker'
    else:
        return 'Surfer'

df['User Type'] = np.vectorize(user_type)(df['Talked'], df['Internet'])
display('4.b. Using Vectorize: ', df)
```

```python
# 5.a. Sort subscribers by talk time from smallest to largest

display('5.a. Suscribers by talked time in ascending order: ', df.sort_values(['Talked'], ascending=True))
print()

# 5.b. Sort subscribers by talk time + browsing time

display('5.b. Suscribers by talked and Internet time in descending order: ', df.sort_values(['Talked', 'Internet'], ascending=False))
```

```python
# 6.a. Print the 5 largest talkers. Hint: nlargest

display('6.a. The 5 suscribers with more talked time: ', df.nlargest(5, 'Talked'))
print()

# 6.b. Print the 3 least active subscribers. Hint: nsmallest

display('6.b. The 3 sucribers with less talked time: ', df.nsmallest(3, 'Talked'))
```

```python
# 7 Some of the records appear to have been recorded multiple times -
# a. Print the duplicate records

display('7.a. Duplicated records: ', df.duplicated())
print()

# b. Remove the duplicate records

display('7.b. Table after removing the duplicated records: ', df.drop_duplicates())
```
